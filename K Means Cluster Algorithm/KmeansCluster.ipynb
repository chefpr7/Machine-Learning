```python
#importing libraries
#ML Classification Model
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from k_means_clustering import K_Means
```

```python
#Loading data
data=np.loadtxt(open('Image2.txt',"rt"),delimiter=",")
X=data[:,:2]
```

```python
#Creating instance of class
kmeans = K_Means(K=2)
X=kmeans.normalize(X)         #Normalizing features    
```

```python
# VISUALIZATION of untrained data
plt.scatter(X[:,0],X[:,1])
plt.xlabel("FEATURE X1")
plt.ylabel("FEATURE X2")
plt.title("UNTRAINED DATA")
plt.show()
```

```python
#Unsupervised learning carried out on dataset X
kmeans.fit(X)
```

```python
# VISUALIZATION of trained data
plt.scatter(X[:,0],X[:,1],c=kmeans.labels)
plt.xlabel("FEATURE X1")
plt.ylabel("FEATURE X2")
plt.title("TRAINED DATA(My algorithm)")
plt.show()
```

```python
# unsupervised learning by sklearn 
kmeans = KMeans(n_clusters=2)  
kmeans.fit(X)
```

```python
# VISUALIZATION of training data according to sklearn
plt.scatter(X[:,0],X[:,1], c=kmeans.labels_)
plt.xlabel("FEATURE X1")
plt.ylabel("FEATURE X2")
plt.title("TRAINED DATA(sklearn)")
plt.show()
```
